```python
import numpy as np
import matplotlib.pyplot as plt

# Funções de Hermite
def H1(t):
    return 2 * t**3 - 3 * t**2 + 1

def H2(t):
    return -2 * t**3 + 3 * t**2

def H3(t):
    return t**3 - 2 * t**2 + t

def H4(t):
    return t**3 - t**2

# Função de interpolação de Hermite
def hermite_interpolation(p0, p1, m0, m1, t):
    return p0 * H1(t) + p1 * H2(t) + m0 * H3(t) + m1 * H4(t)

# Definindo os pontos e tangentes
P1 = np.array([0, 0])  # Ponto inicial
P2 = np.array([3, 3])  # Ponto final
M1 = np.array([6, -6])  # Tangente no ponto inicial
M2 = np.array([6, 4])   # Tangente no ponto final

# Valores de t de 0 a 1, com passo de 0.1 (10 segmentos)
t_values = np.linspace(0, 1, 11)

# Inicializando listas para armazenar os valores de interpolação
curve_points = []

# Gerando a curva para cada valor de t
for t in t_values:
    point = hermite_interpolation(P1, P2, M1, M2, t)
    curve_points.append(point)

# Convertendo a lista de pontos para um array numpy para facilitar o plot
curve_points = np.array(curve_points)

# Plotando a curva
plt.plot(curve_points[:, 0], curve_points[:, 1], label="Curva de Hermite", marker='o')
plt.scatter([P1[0], P2[0]], [P1[1], P2[1]], color="red", label="Pontos de Controle")
plt.quiver(P1[0], P1[1], M1[0], M1[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Tangente no P1")
plt.quiver(P2[0], P2[1], M2[0], M2[1], angles='xy', scale_units='xy', scale=1, color='green', label="Tangente no P2")

# Ajustes do gráfico
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Curva de Hermite com 10 segmentos")
plt.legend()
plt.grid(True)
plt.show()
```
